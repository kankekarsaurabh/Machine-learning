Stemming

Process of reducing word to it word stem


```python
#words
import nltk
```

```python

words=['writes','writing', 'eat', 'eating', 'final', 'finale', 'hist', 'historical']
```

```python
#porter stemmer

from nltk.stem import PorterStemmer
```

```python
stemming= PorterStemmer()
```

```python
for word in words:
  print(word+"---->"+stemming.stem(word))
```

**Regex Stemmer**


A Regex Stemmer is a type of stemming algorithm that utilizes regular expressions to identify and remove morphological affixes (like suffixes or prefixes) from words, thereby approximating their root forms. This method is often used in Natural Language Processing (NLP) to reduce words to a common base for text analysis and information retrieval.

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer=RegexpStemmer('ing|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem("ingeating")
```
